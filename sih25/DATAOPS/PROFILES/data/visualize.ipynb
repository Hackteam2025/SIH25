```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the NetCDF file
# Replace 'your_file.nc' with your actual file path
ds = xr.open_dataset('1900121_meta.nc')

# Basic information about the dataset (similar to df.info())
print("Dataset Overview:")
print(ds)
print("\n" + "="*50 + "\n")

# Get dataset dimensions
print("Dimensions:")
print(ds.dims)
print("\n" + "="*50 + "\n")

# Get coordinate information
print("Coordinates:")
print(ds.coords)
print("\n" + "="*50 + "\n")

# Get data variables
print("Data Variables:")
print(ds.data_vars)
print("\n" + "="*50 + "\n")

# Get attributes (metadata)
print("Global Attributes:")
print(ds.attrs)
```
